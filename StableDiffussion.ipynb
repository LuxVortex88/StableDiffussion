```python
from diffusers import StableDiffusionPipeline
import torch
import os # Import the os module to check for CUDA availability

model_id = "runwayml/stable-diffusion-v1-5"

# Check if CUDA is available
if torch.cuda.is_available():
    device = "cuda"
    dtype = torch.float16
    print("CUDA is available. Using GPU.")
else:
    device = "cpu"
    dtype = torch.float32 # Use float32 for CPU as float16 is not always supported or beneficial
    print("CUDA is not available. Using CPU.")


pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=dtype, # Use the determined dtype
)

# Move the pipe to the determined device
pipe = pipe.to(device)

# Enable attention slicing for lower memory usage if needed
# Attention slicing is generally less effective on CPU, but the method exists
if device == "cuda":
    pipe.enable_attention_slicing()

# Step 3: Generate the image
# Create the generator on the determined device
generator = torch.Generator(device).manual_seed(42)  # for reproducibility

# Pass the prompt to the pipe function
prompt = "a photo of an astronaut riding a horse on the moon" # Example prompt - replace with your desired prompt

image = pipe(prompt, guidance_scale=7.5, num_inference_steps=50, generator=generator).images[0]

# Step 4: Save or display the image
image.save("recursive_root_electron_wave.png")
# image.show() # image.show() requires Pillow and a display environment. Saving is more reliable.
```
